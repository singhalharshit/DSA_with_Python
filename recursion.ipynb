```python
def facto(n):
    if n==1:
        return 1
    smallAnswer = facto(n-1)
    ans = n * smallAnswer
    return ans
    
```

```python
facto(10)
```

```python
def pow2(n):
    if n==1:
        return 2
    
    ans=pow2(n-1)
    true_ans = 2*ans
    return true_ans


pow2(5)
```

```python
# Find the digits in a given number

"""
    234-> 3
    54321-> 5
"""

def count_digits(n):
    if n>0 and n<10:
        return 1
    elif n==0:
        return 0

    smallN = int(n/10)
    smallAns = count_digits(smallN)
    ans = 1+smallAns
    return ans
```

```python
count_digits(1234)
```
