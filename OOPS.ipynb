```python
class Student:
    numberOfStudents =0
    def __init__(self,name,rollNumber,marks):
        self.name = name
        self.rollNumber = rollNumber
        self.__marks = marks #Making Private Variable
        Student.numberOfStudents = Student.numberOfStudents+1
    
    def getMarks(self):
        return self.__marks

    def setMarks(self,newMarks):
        self.__marks = newMarks
        return newMarks
    
    def study(self):
        print("I am studying")
        
    def play(self):
        print("Done with studies now I'll play")
```

```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact



fact(10)
```

```python
def facto(n):
    if n==1:
        return 1
    smallAnswer = fact(n-1)
    ans = n * smallAnswer
    return ans
    
```

```python
# This script prepares a structured list of 400+ Python & SQL interview questions
# (200 real interview + 200 mock) and prepares them for PDF export via FPDF

from fpdf import FPDF
import textwrap

# Define the PDF structure
class InterviewPDF(FPDF):
    def header(self):
        self.set_font("Arial", 'B', 16)
        self.cell(0, 10, "Python + SQL Interview Handbook (Real + Mock Questions)", ln=True, align='C')
        self.ln(5)

    def chapter_title(self, title):
        self.set_font("Arial", 'B', 12)
        self.set_fill_color(220, 220, 220)
        self.cell(0, 10, title, ln=True, fill=True)
        self.ln(2)

    def chapter_body(self, questions):
        self.set_font("Arial", size=11)
        for i, qa in enumerate(questions, 1):
            wrapped_q = textwrap.fill(f"Q{i}: {qa['q']}", width=100)
            wrapped_a = textwrap.fill(f"A: {qa['a']}", width=100)
            self.multi_cell(0, 10, wrapped_q)
            self.multi_cell(0, 10, wrapped_a)
            self.ln(3)
        self.ln(5)

# Sample starter dataset: Extend these lists to 200+ each
real_python_questions = [
    {"q": "What is a Python generator? Explain with use case.", "a": "A generator yields values using 'yield'. Useful for streaming large data."},
    {"q": "What is the GIL in Python and how does it affect threading?", "a": "Global Interpreter Lock allows only one thread at a time, affecting CPU-bound tasks."},
    # ...add 198 more real Python questions
]

mock_python_questions = [
    {"q": "Write a Python function to reverse a string.", "a": "def reverse_string(s): return s[::-1]"},
    {"q": "How does list comprehension work in Python?", "a": "A concise way to create lists: [x for x in iterable if condition]"},
    # ...add 198 more mock Python questions
]

real_sql_questions = [
    {"q": "Write SQL to find the second highest salary.", "a": "SELECT MAX(salary) FROM emp WHERE salary < (SELECT MAX(salary) FROM emp);"},
    {"q": "Explain the difference between RANK() and DENSE_RANK().", "a": "RANK() skips ranks if there's a tie, DENSE_RANK() does not."},
    # ...add 198 more real SQL questions
]

mock_sql_questions = [
    {"q": "Write a query to count employees in each department.", "a": "SELECT dept_id, COUNT(*) FROM emp GROUP BY dept_id;"},
    {"q": "How do you fetch the top 3 records per group in SQL?", "a": "Use ROW_NUMBER() OVER (PARTITION BY ...) and filter WHERE row_num <= 3"},
    # ...add 198 more mock SQL questions
]

# To create the PDF (you can uncomment and run this in a local Python env with FPDF installed)
# pdf = InterviewPDF()
# pdf.set_auto_page_break(auto=True, margin=15)
# pdf.add_page()
# pdf.chapter_title("Real Python Interview Questions")
# pdf.chapter_body(real_python_questions)
# pdf.chapter_title("Mock Python Interview Questions")
# pdf.chapter_body(mock_python_questions)
# pdf.chapter_title("Real SQL Interview Questions")
# pdf.chapter_body(real_sql_questions)
# pdf.chapter_title("Mock SQL Interview Questions")
# pdf.chapter_body(mock_sql_questions)
# pdf.output("Python_SQL_Interview_Handbook.pdf")
```
